````
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

## Importing the dataset


````
dataset = pd.read_csv('/content/drive/MyDrive/ML_Udemy_tutorial/Teacher_Files/part2-Regression/Section 4 _ Simple Linear Regresion/Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
````

## Splitting the dataset into the Training set and Test set

````
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) # or =>  train_test_split(X, y, test_size = 1/3, random_state = 0)
````

## Training the Simple Linear Regression model on the Training set


```
Regression => predict the continus real value like salary
Classification => predict the category or class

this example is about year of experience(X) and salary(y), so we can do this countinus problem with Linear Regresion.
```



````
#create simple regression model:
from sklearn.linear_model import LinearRegression # call Simple Linear Regresion class
regressor=LinearRegression() #create Simple Linear Regression
regressor.fit(X_train,y_train) #train linear regresion on training set(on features(x) and output vector(y)), because to do linear regression we need the features and output result

````



## Predicting the Test set results


```
y_test => is Real Salary
y_pred => is Predicted Salary
```



````
y_pred=regressor.predict(X_test)
````

````
print(X_test)
print(y_pred)
````

## Visualising the Training set results


```
Scatter plot let us to visualize the red point on it
```



````
plt.scatter(X_train,y_train,color='red') #determine the x , y axis
plt.plot(X_train,regressor.predict(X_train),color='blue') #show the curve or line of Linear Regresion in plot. determine the coordinate of line, we want to show the corespond regression output predicted for each X_train
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
````

## Visualising the Test set results

````
plt.scatter(X_test,y_test,color='red') #determine the x , y axis
plt.plot(X_train, regressor.predict(X_train), color = 'blue') #show the curve or line of Linear Regresion in plot. determine the coordinate of line, we want to show the corespond regression output predicted for each X_train
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
````

## Making a single prediction (for example the salary of an employee with 12 years of experience)

````
print(regressor.predict([[12]]))
````

Therefore, our model predicts that the salary of an employee with 12 years of experience is $ 138967,5.

**Important note:** Notice that the value of the feature (12 years) was input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting 12 into a double pair of square brackets makes the input exactly a 2D array. Simply put:

$12 \rightarrow \textrm{scalar}$

$[12] \rightarrow \textrm{1D array}$

$[[12]] \rightarrow \textrm{2D array}$

## Getting the final linear regression equation with the values of the coefficients

````
print(regressor.coef_)
print(regressor.intercept_)
````

Therefore, the equation of our simple linear regression model is:

$$\textrm{Salary} = 9345.94 \times \textrm{YearsExperience} + 26816.19$$

**Important Note:** To get these coefficients we called the "coef_" and "intercept_" attributes from our regressor object. Attributes in Python are different than methods and usually return a simple value or an array of values.

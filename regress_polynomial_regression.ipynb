## Importing the dataset

````
dataset = pd.read_csv('/content/drive/MyDrive/ML_Udemy_tutorial/Teacher_Files/part2-Regression/Section 6 _ Ploynomial Linear Regression/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values #i want all columns except 0 and last column
y = dataset.iloc[:, -1].values
````

## Training the Linear Regression model on the whole dataset


```
at the first we shoud creat linear regression and then we will create the polynomial Regression
```



````
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
````

## Training the Polynomial Regression model on the whole dataset


```
-  X-poly (matrix of feature)=> first we hve to create matrix of features(x1,x1 ^ 2 ,...x1^n)
- then we create linear regression class to integrate power feature of matrix feature (y=b0+b1x1+b2X1^2,..)
```



````
from sklearn.preprocessing import PolynomialFeatures # this produce x1,x1^2 ,..x1^n
poly_reg=PolynomialFeatures(degree=4) #create object of plolynomial class# degree is n (which in this example n is the position level that is 4 and y is the salary)
X_poly= poly_reg.fit_transform(X) #x is level column and x_poly is=> x1,x1^2 ,x1^3,x1^4

#to add above matrix of feature(X_poly) to the equision we should use of anothr linear regression:
lin_reg_2=LinearRegression()
lin_reg_2.fit(X_poly,y)
````

## Visualising the Linear Regression results

````
plt.scatter(X,y,color='red') #ground truth
plt.plot(X,lin_reg.predict(X),color='blue') #predicted linear regression
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
````

## Visualising the Polynomial Regression results


```
underfitting=> when  predicted values is simpler than the ground_truth data . so I must add new power to the x like x^2,x^3,x^n (and in deep learning we nust add new layer)

overfitting=> when my model be more precise than the ground_truth (lern noisy data as well)( so we should use grid search or  dropout in deep learning and something like this)
```



````
plt.scatter(X,y,color='red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)),color='blue') #predicted polynomial regression
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
````

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

````
X_grid = np.arange(min(X), max(X), 0.1) #0.1 to show the curve  smoothly (instead of 1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
````
